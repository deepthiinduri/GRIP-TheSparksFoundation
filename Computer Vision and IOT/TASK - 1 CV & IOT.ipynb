### Importing libraries

```python
import matplotlib.pyplot as plt
import pytesseract 
import easyocr
import cv2
import numpy as np
import matplotlib.image as mpimg
%matplotlib inline

try:
    from PIL import Image
except ImportError:
    import Image
```

### Example 1

```python
img1 = mpimg.imread('image1.png')
imgplot = plt.imshow(img1)

# Show image
plt.show()
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\DELL\AppData\Local\Tesseract-OCR\tesseract.exe' #changing path
```

```python
def recText(filename):
    text = pytesseract.image_to_string(Image.open(filename))
    return text
```

```python
# Import image
info1 = recText('image1.png')
print(info1) #printing text on the image

file = open("result1.txt","w")
file.write(info1)
file.close()
print("\n\nWritten successful")
```
